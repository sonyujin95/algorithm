```python
class Solution:
    def solution(self, number, dictionary):
        # 알파벳 순으로 정렬
        dictionary.sort()
        result = []
        
        for n in number:
            n = int(n)
            for x in dictionary:
                # 단어의 길이와 숫자가 같다면 result에 추가하고 해당 단어를 제거한다.
                if len(x) == n:
                    result.append(x)
                    dictionary.pop(dictionary.index(x))
                    break
        return ' '.join(result)
```
