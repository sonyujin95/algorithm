```python
def solution(d, budget):
    # d(부서별 신청 금액)를 정렬한다.
    d.sort()
    # 전체 부서별 신청금액보다 예산이 크면
    while budget < sum(d):
        # 가장 큰 금액을 지원한 부서부터 하나씩 값을 뺀다.
        d.pop()
    # 모든 부서를 지원할 수 있을 때 부서의 수를 리턴한다.
    return len(d)
```

```python
# 입출력 예 1
print(solution([1,3,2,5,4],9))
```

```python
# 입출력 예 2
print(solution([2,2,2,3],10))
```
