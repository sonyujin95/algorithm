```python
def solution(numbers):
    from itertools import permutations       # itertools  (x) 
    import math                               # math 
    num=[i for i in numbers]                  #  

    array=[]                                   # array list
    for i in range(1,len(num)+1):
        array+=list(permutations(num,i))       #   .

    lst=[]                                     # lst list
    for i in array:                           # array  
        temp=''
        for j in range(len(i)):
            temp+=i[j]
        lst.append(int(temp))
    set_array=set(lst)                         #  
    
    def check(x):                                  
        if x<2:
            return False                           # 2  x(0 1  )

        for i in range(2, int(math.sqrt(x))+1):
            if x%i==0:
                return False                       #   x
        return True                                #       .

    result=[]                                        #  list
    for i in set_array:
        if check(i):
            result.append(i)                         #   
    return len(result)                              #   
```
