```python
def solution(numbers):
    from itertools import permutations       # itertools 라이브러리로부터 순열(중복x) 불러오기
    import math                               # math 불러오기
    num=[i for i in numbers]                  # 숫자에서 반복

    array=[]                                   # array list
    for i in range(1,len(num)+1):
        array+=list(permutations(num,i))       # 순열리스트에서 문자열을 합쳐준다.

    lst=[]                                     # lst list
    for i in array:                           # array로부터 순열리스트를 받음
        temp=''
        for j in range(len(i)):
            temp+=i[j]
        lst.append(int(temp))
    set_array=set(lst)                         # 중복값 제거
    
    def check(x):                                  
        if x<2:
            return False                           # 2 밑이면 x(0과 1은 소수가 아님)

        for i in range(2, int(math.sqrt(x))+1):
            if x%i==0:
                return False                       # 소수가 아니면 x
        return True                                # 소수가 아닌 것을 다 걸러서 소수만 남는다.

    result=[]                                        # 결과값 list
    for i in set_array:
        if check(i):
            result.append(i)                         # 소수면 개수 추가
    return len(result)                              # 소수 개수 출력
```
