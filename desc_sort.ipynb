```python
def solution(s):
    answer = ''
    return (answer.join(sorted(s)[::-1]))  # sort로 s를 정렬하고 -1로 순서를 뒤집어준 뒤, join으로 연결한다.
```

```python
# 입출력 예 확인
s = 'Zbcdefg'
print(solution(s))
```
