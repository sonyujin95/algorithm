```python
def solution(a, b):
    if a == b:
        return a    #     a 
    
    if a >= b:
        max_value = a
        min_value = b  # a    a , b 
        
    else:              #  a     b  
        max_value = b
        min_value = a   #  b , a 
        
    result = 0          # 
    for i in range(min_value, max_value + 1):
        result += i     #  (+1     )    
        
    return result      #  
```
