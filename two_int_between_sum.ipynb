```python
def solution(a, b):
    if a == b:
        return a    # 두 정수 값이 같으면 a값을 리턴
    
    if a >= b:
        max_value = a
        min_value = b  # a가 더 크거나 같으면 a값이 최댓값, b값이 최솟값
        
    else:              # 같을때와 a가 클 때를 했으니 나머지인 b가 클 때
        max_value = b
        min_value = a   #  b가 최댓값, a가 최솟값
        
    result = 0          # 초기화
    for i in range(min_value, max_value + 1):
        result += i     # 최솟값과 최댓값(+1을 안해주면 최댓값 이전 값이 나옴)사이 값들을 더해서 나온 값
        
    return result      # 값 출력
```
