```python
def rotate_matrix_90(one):
    N = len(one)
    res = [[0] * N for _ in range(N)]

    for c in range(3):
        for row in range(N):
            for col in range(N):
                res[col][N-1-row] = one[row][col]
        return res
```

```python
cone_90 = np.array(rotate_matrix_90(one))

print(cone_90)
```

```python
cv2.imshow('color', cone_90)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def rotate_matrix_180(one):
    N = len(one)
    res = [[0] * N for _ in range(N)]    # 변형할 행렬

    for c in range(3):
        for row in range(N):
            for col in range(N):
                res[N - 1 - row][N - 1 - col] = one[row][col]
    return res
```

```python
cone_180 = np.array(rotate_matrix_180(one))
print(cone_180)
```

```python
cv2.imshow('180', cone_180)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def rotate_matrix_270(one):
    N = len(one)
    res = [[0] * N for _ in range(N)]   # 변형할 행렬

    for c in range(3):
        for row in range(N):
            for col in range(N):
                res[N - 1 - col][row] = one[row][col]
    return res
```

```python
cone_270 = np.array(rotate_matrix_270(one))
print(cone_270)
```

```python
cv2.imshow('270', cone_270)
cv2.waitKey()
cv2.destroyAllWindows()
```
