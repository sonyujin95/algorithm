```python
def solution(n):
    new = str(n)    # n을 문자열로 바꿔서 재정의
    answer = 0      # 다 더한값 0으로 초기화
     
    for i in range(len(new)):     # 믄지열길이만큼 반복
        answer += int(new[i])     # 문자열을 정수로 바꿔 각 자리수마다 자리수 끝날때까지 더하기 반복
    return answer                 # 각 자리수 더한값 반환


# 입출력 예 직접 넣어서 출력하기
print(solution(123))
print(solution(987))
```
