```python
def solution(array, commands):
    answer = []
    # commands에 있는 command(i, j, k)를 뽑는다.
    for command in commands:
        i, j, k = command[0], command[1], command[2]
        slice = array[i-1:j]  # array에서 slice를 한다.
        slice.sort()  # slice한 것을 정렬한다.
        answer.append(slice[k-1])  # slice한 것을인덱싱한다.
    return answer
```
