```python
def solution(array, commands):
    answer = []
    # commands  command(i, j, k) .
    for command in commands:
        i, j, k = command[0], command[1], command[2]
        slice = array[i-1:j]  # array slice .
        slice.sort()  # slice  .
        answer.append(slice[k-1])  # slice .
    return answer
```
