```python
def gcd(n1, n2):
    if n1 < n2:              # n2 가 n1보다 크면
        (n1, n2) = (n2, n1)  # 둘의 값을 바꿔준다.
        
    while n2 != 0:          # n2 가 0이 아니라면
        (n1, n2) = (n2, n1 % n2)  # 둘의 자리를 바꾼 후 n1을 n2로 나눈다
        
    return n1               # n1값을 반환한다

def solution(n, m):
    return [gcd(n, m), (n * m) / gcd(n, m)]    # 최소공배수는 n과 m을 곱한 후 최대공약수로 나눠준다
```
