```python
def solution(bridge_length, weight, truck_weights):
    stack = [0] * bridge_length        # 다리 길이만큼 리스트를 생성한다.
    answer = 0
    while stack:
        stack.pop(0)        # 리스트의 맨 앞의 값을 가져온다.
        answer+=1           # 1초가 지날 때마다 스택을 한 칸씩 왼쪽으로 미뤄준다.
        if truck_weights:
            if (sum(stack)+truck_weights[0])<=weight:    # 이미 다리에 올라간 트럭의 무게와 다음에 올라갈 트럭의 무게 합이 다리가 수용한 무게라면
                stack.append(truck_weights.pop(0))        # 스택에 추가한다.
            else:                                        # 그렇지 않다면
                stack.append(0)                           # 0을 추가해 밀어낸다.
    return answer
```
