```python
def solution(s):
    if (len(s) == 4 or len(s) == 6) and s.isdigit():   # 문자열 길이가 4이거나 6이고, 숫자이면
        return True                                    # True 반환
    else:                                               # 그렇지 않다면
        return False                                   # False 반환

# 입출력 예 test
s="1234"
print(solution(s))
```

```python
# 입출력 예 test
s="a234"
print(solution(s))
```
