```python
def solution(n, lost, reserve):
    # reserve의 요소들 중 lost에 동일하게 존재하는 값들을 모두 제거한다.
    reserve_uniq = set(reserve) - set(lost)
    lost_uniq = set(lost) - set(reserve)
    
    for i in reserve_uniq:
        # reverse학생의 왼쪽 학생부터 빌려준다.
        if i-1 in lost_uniq:
            lost_uniq.remove(i-1)
        # 만약에 없다면 오른쪽 학생을 빌려준다.
        elif i+1 in lost_uniq:
            lost_uniq.remove(i+1)
            
    return n - len(lost_uniq)
```

```python
# 입출력 예 1
print(solution(5, [2,4],[1,3,5]))
```

```python
# 입출력 예 2
print(solution(5, [2,4], [3]))
```

```python
# 입출력 예 3
print(solution(3, [3], [1]))
```
