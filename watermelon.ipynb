```python
def solution(n):
    answer = ''
    
    for i in range(n):              # n 
        if i % 2 == 0:              #  i 
            answer += ''          #  
            
        else:                       # 
            answer += ''          #  
            
    return answer


# n = 3 i = 0, answer = ''/ i = 1 , answer = '' / i = 2, answer = ''
# n = 4 i = 0, answer = ''/ i = 1 , answer = '' / i = 2, answer = '' / i = 3, answer = ''
```
