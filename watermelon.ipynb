```python
def solution(n):
    answer = ''
    
    for i in range(n):              # n번만큼 반복
        if i % 2 == 0:              # 만약 i가 짝수이면
            answer += '수'          # 수 추가
            
        else:                       # 홀수이면
            answer += '박'          # 박 추가
            
    return answer


# n = 3이면 i = 0, answer = '수'/ i = 1 , answer = '수박' / i = 2, answer = '수박수'
# n = 4이면 i = 0, answer = '수'/ i = 1 , answer = '수박' / i = 2, answer = '수박수' / i = 3, answer = '수박수박'
```
