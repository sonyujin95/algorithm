```python
def solution(answers):
    stu1 = [1, 2, 3, 4, 5] * 2000
    stu2 = [2, 1, 2, 3, 2, 4, 2, 5] * 1250
    stu3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5] * 1000
    
    correct = [0, 0, 0]                # [stu1, stu2, stu3] 순서대로 리스트에 출력
    for i in range(len(answers)):
        if answers[i] == stu1[i]:      # 반복문을 통해 1번학생 답이 맞으면
            correct[0] += 1            # 0번째에 1을 추가
        if answers[i] == stu2[i]:      # 2번학생 답이 맞으면
            correct[1] += 1            # 1번째에 1을 추가
        if answers[i] == stu3[i]:      # 3번학생 답이 맞으면
            correct[2] += 1            # 2번째에 1을 추가
            
    array = []
    for i in range(3):
        array.append((correct[i], i-1))  
        
    array.sort(key = lambda x : (-x[0]))
    
    result = []
    temp = 0
    for i in array:
        value, index = i
        if value >= temp:
            result.append(index)
            temp = value                # 두 값이 같을때까지 계속 추가해서 반복
    return result                      # 해당하는 수포자(인덱스) 출력
```
