```python
def solution(answers):
    stu1 = [1, 2, 3, 4, 5] * 2000
    stu2 = [2, 1, 2, 3, 2, 4, 2, 5] * 1250
    stu3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5] * 1000
    
    correct = [0, 0, 0]                # [stu1, stu2, stu3]   
    for i in range(len(answers)):
        if answers[i] == stu1[i]:      #   1  
            correct[0] += 1            # 0 1 
        if answers[i] == stu2[i]:      # 2  
            correct[1] += 1            # 1 1 
        if answers[i] == stu3[i]:      # 3  
            correct[2] += 1            # 2 1 
            
    array = []
    for i in range(3):
        array.append((correct[i], i-1))  
        
    array.sort(key = lambda x : (-x[0]))
    
    result = []
    temp = 0
    for i in array:
        value, index = i
        if value >= temp:
            result.append(index)
            temp = value                #      
    return result                      #  () 
```
