```python
def solution(n, arr1, arr2):
    answer = []  # 리스트 answer를 생성한다.
    for i,j in zip(arr1,arr2):
        # arr1과 arr2의 값을 비트연산자 or(|)를 이용해 구하고, [2:]를 통해 0b를 제거한다.
        a12 = str(bin(i|j)[2:])
        # a12를 오른쪽 정렬해서 n보다 길이가 작으면 앞에는 0으로 채운다.
        a12=a12.rjust(n,'0')  
        a12=a12.replace('1','#')  # 1이면 #으로 바꿔준다.
        a12=a12.replace('0',' ')  # 0이면 공백으로 바꿔준다.
        answer.append(a12)  # 리스트 answer에 저장한다.
    return answer  # 리스트 answer를 반환한다.
```

```python
# 입출력 에 1
print(solution(5, [9,20,28,18,11], [30, 1, 21, 17, 28]))
```

```python
# 입출력 에 2
print(solution(6, [46,33,33,22,31, 50], [27, 56, 19, 14, 14, 10]))
```
