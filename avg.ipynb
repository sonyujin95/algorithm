```python
def solution(arr):
    answer = 0            # answer 초기값
    for i in arr:         # arr(리스트) 안에 i 반복
        answer = answer + i  # answer에 i값 추가
    return answer / len(arr)   # 평균값 도출 (리스트내 합/리스트개수)


# 입출력 예
arr = [1, 2, 3, 4]
print("평균값 : {}".format(solution(arr)));
```

```python
arr = [5, 5]
print("평균값 : {}".format(solution(arr)));
```
