```python
def solution(arr1, arr2):
    answer = [[] for x in range(len(arr2))]     #    []   arr2   
    for i in range(len(arr1)):                  #  
        for j in range(len(arr1[i])):           #  
            answer[i].append(arr1[i][j] + arr2[i][j])   #    i        .
            
            
    return answer
```
