```python
def solution(arr1, arr2):
    answer = [[] for x in range(len(arr2))]     # 빈 리스트를 만들고 [] 의 개수를 arr2의 길이와 같게 설정한다
    for i in range(len(arr1)):                  # 세로 크기
        for j in range(len(arr1[i])):           # 가로 크기
            answer[i].append(arr1[i][j] + arr2[i][j])   # 빈 리스트의 인덱스 i에 각 행렬의 해당 인덱스 값을 더한 값을 붙여준다.
            
            
    return answer
```
