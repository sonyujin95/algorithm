```python
def solution(s, n):
    s = list(s)  #  s 
    for i in range(len(s)):
        if s[i].isupper():  # s  
            # chr =   -> , ord=  ->  
            s[i]=chr((ord(s[i])-ord('A')+n)%26+ord('A'))  #   n   
        elif s[i].islower():  # 
            s[i]=chr((ord(s[i])-ord('a')+n)%26+ord('a'))  #   n   
    return "".join(s)
```

```python
#   1
print(solution("AB",1))
```

```python
#   2
print(solution("z",1))
```

```python
#   3
print(solution("a B z",4))
```
