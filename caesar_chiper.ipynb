```python
def solution(s, n):
    s = list(s)  # 리스트 s 생성
    for i in range(len(s)):
        if s[i].isupper():  # 리스트s의 요소가 대문자이면
            # chr = 아스키 코드 -> 문자, ord= 문자 -> 아스키 코드
            s[i]=chr((ord(s[i])-ord('A')+n)%26+ord('A'))  # 대문자에 대하여 n의 숫자만큼 밀게 하고
        elif s[i].islower():  # 소문자이면
            s[i]=chr((ord(s[i])-ord('a')+n)%26+ord('a'))  # 소문자에 대하여 n의 숫자만큼 밀게 한다
    return "".join(s)
```

```python
# 입출력 예 1
print(solution("AB",1))
```

```python
# 입출력 예 2
print(solution("z",1))
```

```python
# 입출력 예 3
print(solution("a B z",4))
```
