```python
def solution(arr):
    answer = []      # list로 생성
    answer.append(arr[0])     # list 첫번째 값 입력
    for i in range(1, len(arr)):   # 1부터 arr길이까지 반복
        if arr[i] != arr[i-1]:     # arr[i] 값과 그 전값이 다르다면
            answer.append(arr[i])   # arr[i] 에 추가
    
    # [실행] 버튼을 누르면 출력 값을 볼 수 있습니다.
    print('Hello Python')
    return answer
```
