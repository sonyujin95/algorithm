```python
def solution(arr, divisor):
    answer = []  #  
    for i in range(len(arr)):  # 
        if arr[i] % divisor == 0:  # arr [i] divisor  
            answer.append(arr[i])  # answer arr[i] 
    if len(answer) == 0:   #    
        answer.append(-1)  # answer -1 
    else:  # 
        answer.sort()  # answer .
    return answer  # answer .
```
