```python
def solution(arr, divisor):
    answer = []  # 리스트로 정의
    for i in range(len(arr)):  # 
        if arr[i] % divisor == 0:  # arr리스트의 [i]요소가 divisor와 나누어 떨어지면
            answer.append(arr[i])  # answer리스트에 arr[i]를 추가
    if len(answer) == 0:   # 만약 나누어떨어지는 요소가 없다면
        answer.append(-1)  # answer에 -1을 추가
    else:  # 있다면
        answer.sort()  # answer를 정렬한다.
    return answer  # answer를 반환한다.
```
