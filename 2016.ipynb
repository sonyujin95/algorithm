```python
def solution(a, b):
    # 2016년의 시작은 금요일이다.
    date = ['FRI', 'SAT', 'SUN', 'MON', 'TUE', 'WED', 'THU']
    # 2월은 윤년이라 29일까지 있다.
    month = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    # 요일 반환, a는 월, b는 일이다. 7로 나눈 나머지값의 date 위치가 곧 요일이다.
    return date[(sum(month[:a-1])+b-1)%7]
```

```python
# 입출력 예
print(solution(5,24))
```
